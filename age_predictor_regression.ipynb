```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, f1_score
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, GaussianNoise
from keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.initializers import RandomNormal
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset.csv')
df.head()
```

```python
# One-hot della colonna con stringhe e passaggio a numpy

# df = pd.get_dummies(df, columns=['materiale'])
db = df.to_numpy().astype(float)

# Seleziona la colonna dei label e la toglie dal dataset

X = db[:,1:]
y = db[:,0].astype(str)
num_classes = len(np.unique(y))

print(set(y))

y = y.astype(float)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = RandomForestRegressor(criterion='squared_error',max_features=None,n_estimators=100, n_jobs = -1,random_state=42,verbose=1,min_samples_split=50)

model.fit(X_train,y_train)

y_pred = model.predict(X_test)
```

```python
score_abs = mean_absolute_error(y_test.astype(float), y_pred.astype(float))
```

```python
print(score_abs)
```

```python
keras.utils.set_random_seed(42)

model = Sequential()
model.add(Dense(1000, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(1, activation='linear'))

loss = 'mean_squared_error'

opt = Adam(learning_rate=1e-3)
model.compile(loss=loss, optimizer=opt)

model.fit(X_train, y_train, epochs=60, batch_size=32, verbose=1)

y_pred = model.predict(X_test)
```

```python
score_abs = mean_absolute_error(y_test, y_pred)
```

```python
print(score_abs)
```
